find the largest value in a list or sequence

```python
largest = None
print('Before', largest)
numbers = ['47', '37', '23', '29', '39', '68']

for itervar in numbers:
    if largest is None or itervar > largest:
        largest = itervar
        print('Loop', itervar, largest)
print('Largest', largest)
```

find the largest value in a list or sequence

```python
smallest = None
print('Before', smallest)
numbers = [67, 40, 34, 59, 50, 12, 54, 57]

for itemsnums in numbers:
    if smallest is None or itemsnums < smallest:
        smallest = itemsnums
        print('Looping', itemsnums, smallest)
print('Smallest: ', smallest)
```
